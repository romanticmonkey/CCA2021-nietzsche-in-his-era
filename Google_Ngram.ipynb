# Downloading files from Google API site

```python
url = 'http://storage.googleapis.com/books/ngrams/books/datasetsv2.html'
soup = BeautifulSoup(rq.get(url).text, 'html.parser')
```

```python
pTags = soup.body.findAll('p')
link_tags = pTags[5].findAll('p')
soup_5gram = link_tags[13]
```

```python
file_url_list = [aTag.get('href') for aTag in soup_5gram.findAll('a')]
```

```python
file_url_list
```

```python
def download(url, folder_dir):
    filename = url[url.rfind('/')+1:]
    desig_dir = folder_dir + filename
    urllib.request.urlretrieve(url, desig_dir)
    print(f'{filename} downloaded successfully.')
```

```python
# Turns out these files are too big for my laptop, might consider cloud solutions in the future

folder_dir = 'data/google_ngrams/'
for url in file_url_list[39:]:
    download(url, folder_dir)
```
